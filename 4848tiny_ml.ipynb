```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from tqdm.notebook import tqdm
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Dense, Conv2D, Dropout, MaxPooling2D, 
                                      Input, BatchNormalization, DepthwiseConv2D, 
                                      GlobalAveragePooling2D, Activation)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from PIL import Image
import cv2

# ===============================
# Configuration - ULTRA LIGHTWEIGHT
# ===============================
BASE_DIR = r"C:\Users\Dishang\OneDrive\Desktop\UTKFace"
IMG_SIZE = (48, 48)  # Reduced to 48x48 for Arduino memory constraints
BATCH_SIZE = 32
EPOCHS = 50

print(f"ğŸ¯ Training ULTRA-LIGHTWEIGHT model for Arduino")
print(f"   Input size: {IMG_SIZE}")
print(f"   Target: <80KB model, <100KB RAM usage")

# ===============================
# Load Dataset with Face Detection
# ===============================
def is_valid_face_image(image_path):
    try:
        img = cv2.imread(image_path)
        if img is None:
            return False
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        return len(faces) > 0
    except:
        return False

print("Loading dataset...")
image_paths = []
age_labels = []
gender_labels = []

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

for filename in tqdm(os.listdir(BASE_DIR)):
    try:
        image_path = os.path.join(BASE_DIR, filename)
        temp = filename.split('_')
        age = int(temp[0])
        gender = int(temp[1])
        
        if is_valid_face_image(image_path):
            image_paths.append(image_path)
            age_labels.append(age)
            gender_labels.append(gender)
    except:
        continue

df = pd.DataFrame()
df['image'], df['age'], df['gender'] = image_paths, age_labels, gender_labels

print(f"âœ“ Total valid face images: {len(df)}")
print(f"âœ“ Gender distribution:\n{df['gender'].value_counts()}")

gender_dict = {0: 'Male', 1: 'Female'}

# ===============================
# Feature Extraction
# ===============================
def extract_features_with_face_focus(images, target_size=IMG_SIZE):
    features = []
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    for image_path in tqdm(images):
        try:
            img = cv2.imread(image_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.1, 4)
            
            if len(faces) > 0:
                x, y, w, h = max(faces, key=lambda f: f[2] * f[3])
                padding = int(0.2 * max(w, h))
                x1 = max(0, x - padding)
                y1 = max(0, y - padding)
                x2 = min(img.shape[1], x + w + padding)
                y2 = min(img.shape[0], y + h + padding)
                img = img[y1:y2, x1:x2]
            
            img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)
            features.append(img)
        except:
            continue
    
    return np.array(features)

print("Extracting features...")
X = extract_features_with_face_focus(df['image'], target_size=IMG_SIZE)
print(f"âœ“ Feature shape: {X.shape}")

# Normalize to [-1, 1] for INT8 quantization
X = X.astype(np.float32) / 255.0
X = X * 2.0 - 1.0

y_gender = np.array(df['gender'][:len(X)])

# Balance classes
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_gender), y=y_gender)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"âœ“ Class weights: {class_weight_dict}")

# Split
X_train, X_test, y_gender_train, y_gender_test = train_test_split(
    X, y_gender, test_size=0.2, random_state=42, stratify=y_gender
)

print(f"âœ“ Training: {X_train.shape}, Test: {X_test.shape}")

# ===============================
# ULTRA-LIGHTWEIGHT Model (Optimized for Arduino)
# ===============================
def create_tiny_gender_model(input_shape=(48, 48, 3)):
    """
    Ultra-lightweight model for Arduino Nano 33 BLE
    Target: <30K parameters, <80KB size, <100KB RAM
    """
    inputs = Input(shape=input_shape, name='input')
    
    # Block 1 - Initial feature extraction
    x = Conv2D(16, (3, 3), strides=(2, 2), padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    # Now: 24x24x16
    
    # Block 2 - Depthwise separable
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(32, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    # Now: 12x12x32
    
    # Block 3 - Depthwise separable
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(64, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    # Now: 6x6x64
    
    # Global pooling
    x = GlobalAveragePooling2D()(x)
    # Now: 64
    
    # Classifier - Very small dense layers
    x = Dense(32, activation='relu')(x)
    x = Dropout(0.5)(x)
    
    # Output
    outputs = Dense(1, activation='sigmoid', name='gender_output')(x)
    
    return Model(inputs, outputs, name='TinyGenderModel')

# Create model
model = create_tiny_gender_model(input_shape=(48, 48, 3))

# Compile
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

model.summary()
total_params = model.count_params()
print(f"\nâœ“ Total Parameters: {total_params:,} (~{total_params/1000:.1f}K)")

# ===============================
# Callbacks
# ===============================
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_tiny_gender_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# ===============================
# Training
# ===============================
print("\nğŸš€ Starting training...")
history = model.fit(
    X_train, y_gender_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_gender_test),
    callbacks=callbacks,
    class_weight=class_weight_dict,
    verbose=1
)

print(f"\nâœ“ Training completed! Epochs: {len(history.history['loss'])}")

# ===============================
# Evaluation
# ===============================
loss, acc, precision, recall = model.evaluate(X_test, y_gender_test, verbose=0)
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"\n{'='*50}")
print(f"KERAS MODEL PERFORMANCE")
print(f"{'='*50}")
print(f"Accuracy:  {acc:.4f} ({acc*100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1_score:.4f}")

# ===============================
# Convert to TensorFlow Lite (INT8 Quantization)
# ===============================
print("\nğŸ”„ Converting to TensorFlow Lite with INT8 quantization...")

def representative_dataset():
    for i in range(min(100, len(X_train))):
        yield [X_train[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8
converter.representative_dataset = representative_dataset

tflite_model = converter.convert()

# Save
tflite_model_path = 'gender_model_tiny_48x48_int8.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

model_size_kb = len(tflite_model) / 1024
print(f"âœ“ TFLite model saved: {tflite_model_path}")
print(f"âœ“ Model size: {model_size_kb:.2f} KB")

# ===============================
# Test TFLite Model & Estimate RAM
# ===============================
print("\nğŸ“Š Testing TFLite model...")
interpreter = tf.lite.Interpreter(model_path=tflite_model_path)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print(f"\nInput Details:")
print(f"  Shape: {input_details[0]['shape']}")
print(f"  Type: {input_details[0]['dtype']}")
print(f"  Size: {np.prod(input_details[0]['shape'])} bytes")

print(f"\nOutput Details:")
print(f"  Shape: {output_details[0]['shape']}")
print(f"  Type: {output_details[0]['dtype']}")

# Test accuracy
correct = 0
total = min(100, len(X_test))

for i in range(total):
    input_scale, input_zero_point = input_details[0]['quantization']
    test_image = X_test[i:i+1].astype(np.float32)
    test_image_quantized = (test_image / input_scale + input_zero_point).astype(np.int8)
    
    interpreter.set_tensor(input_details[0]['index'], test_image_quantized)
    interpreter.invoke()
    
    output_data = interpreter.get_tensor(output_details[0]['index'])
    output_scale, output_zero_point = output_details[0]['quantization']
    output_dequantized = (output_data.astype(np.float32) - output_zero_point) * output_scale
    
    prediction = 1 if output_dequantized[0][0] >= 0.5 else 0
    if prediction == y_gender_test[i]:
        correct += 1

tflite_acc = correct / total
print(f"\nâœ“ TFLite Accuracy (100 samples): {tflite_acc:.4f} ({tflite_acc*100:.2f}%)")

# ===============================
# Estimate Arduino RAM Usage
# ===============================
input_tensor_size = np.prod(input_details[0]['shape'])  # 48*48*3 = 6912 bytes
output_tensor_size = np.prod(output_details[0]['shape'])  # ~1 byte
estimated_arena = model_size_kb * 1.2  # Rough estimate

print(f"\n{'='*50}")
print(f"ARDUINO RAM ESTIMATION")
print(f"{'='*50}")
print(f"Input tensor:     {input_tensor_size:,} bytes ({input_tensor_size/1024:.1f} KB)")
print(f"Output tensor:    {output_tensor_size} bytes")
print(f"Model size:       {model_size_kb:.2f} KB")
print(f"Estimated arena:  {estimated_arena:.0f} KB")
print(f"Total estimated:  {estimated_arena + input_tensor_size/1024:.0f} KB")
print(f"{'='*50}")

if estimated_arena + input_tensor_size/1024 < 120:
    print("âœ… Should fit in Arduino Nano 33 BLE (256KB RAM)")
else:
    print("âš ï¸  May be tight on Arduino Nano 33 BLE memory")

# ===============================
# Convert to C Array
# ===============================
print("\nğŸ“ Generating C header file...")

def convert_to_c_array(tflite_model, header_file_name):
    c_array = ','.join([str(byte) for byte in tflite_model])
    
    header_content = f"""// Auto-generated TensorFlow Lite model
// Model: TINY Gender Classification (48x48, INT8 quantized)
// Generated: {pd.Timestamp.now()}
// Size: {len(tflite_model)} bytes ({len(tflite_model)/1024:.2f} KB)
// Parameters: {total_params:,}

#ifndef GENDER_MODEL_ARDUINO_TINY_H
#define GENDER_MODEL_ARDUINO_TINY_H

const unsigned char gender_model_arduino[] = {{
{c_array}
}};

const unsigned int gender_model_arduino_len = {len(tflite_model)};

#endif // GENDER_MODEL_ARDUINO_TINY_H
"""
    
    with open(header_file_name, 'w') as f:
        f.write(header_content)
    
    print(f"âœ“ Header saved: {header_file_name}")

convert_to_c_array(tflite_model, 'gender_model_arduino_tiny_48x48.h')

# ===============================
# Plot Training History
# ===============================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history.history['accuracy'], label='Train')
axes[0].plot(history.history['val_accuracy'], label='Validation')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Model Accuracy')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history.history['loss'], label='Train')
axes[1].plot(history.history['val_loss'], label='Validation')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].set_title('Model Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# ===============================
# Final Summary
# ===============================
print(f"\n{'='*60}")
print(f"âœ… TRAINING COMPLETE - ULTRA-LIGHTWEIGHT MODEL")
print(f"{'='*60}")
print(f"Parameters:          {total_params:,} (~{total_params/1000:.1f}K)")
print(f"Model Size:          {model_size_kb:.2f} KB")
print(f"Input Shape:         48x48x3 (INT8)")
print(f"Keras Accuracy:      {acc:.4f} ({acc*100:.2f}%)")
print(f"TFLite Accuracy:     {tflite_acc:.4f} ({tflite_acc*100:.2f}%)")
print(f"Estimated RAM:       ~{estimated_arena:.0f} KB")
print(f"{'='*60}")
print(f"\nğŸ“ Files generated:")
print(f"   âœ“ {tflite_model_path}")
print(f"   âœ“ gender_model_arduino_tiny_48x48.h")
print(f"   âœ“ best_tiny_gender_model.h5")
print(f"\nğŸš€ Next: Update Arduino code to use 48x48 input!")
```
