```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from tqdm.notebook import tqdm
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Dense, Conv2D, Dropout, MaxPooling2D, 
                                      Input, BatchNormalization, DepthwiseConv2D, 
                                      GlobalAveragePooling2D, Activation, Add)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from PIL import Image
import cv2

# ===============================
# Configuration
# ===============================
BASE_DIR = r"C:\Users\Dishang\OneDrive\Desktop\UTKFace"
IMG_SIZE = (96, 96)  # Changed to 96x96 for better Arduino compatibility
BATCH_SIZE = 32
EPOCHS = 60

# ===============================
# Load Dataset with Face Detection Validation
# ===============================
def is_valid_face_image(image_path):
    """Validate that image contains a detectable face"""
    try:
        img = cv2.imread(image_path)
        if img is None:
            return False
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        return len(faces) > 0
    except:
        return False

print("Loading dataset...")
image_paths = []
age_labels = []
gender_labels = []

# Load Haar Cascade for validation
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

for filename in tqdm(os.listdir(BASE_DIR)):
    try:
        image_path = os.path.join(BASE_DIR, filename)
        temp = filename.split('_')
        age = int(temp[0])
        gender = int(temp[1])
        
        # Only include images with detectable faces
        if is_valid_face_image(image_path):
            image_paths.append(image_path)
            age_labels.append(age)
            gender_labels.append(gender)
    except:
        continue

df = pd.DataFrame()
df['image'], df['age'], df['gender'] = image_paths, age_labels, gender_labels

print(f"Total valid face images: {len(df)}")
print(f"Gender distribution:\n{df['gender'].value_counts()}")

gender_dict = {0: 'Male', 1: 'Female'}

# ===============================
# Enhanced Data Augmentation
# ===============================
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.2),
])

# ===============================
# Feature Extraction with Preprocessing
# ===============================
def extract_features_with_face_focus(images, target_size=IMG_SIZE):
    """Extract and preprocess face images with face detection"""
    features = []
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    for image_path in tqdm(images):
        try:
            # Load image
            img = cv2.imread(image_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            # Detect face and crop
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.1, 4)
            
            if len(faces) > 0:
                # Crop to largest face
                x, y, w, h = max(faces, key=lambda f: f[2] * f[3])
                # Add padding
                padding = int(0.2 * max(w, h))
                x1 = max(0, x - padding)
                y1 = max(0, y - padding)
                x2 = min(img.shape[1], x + w + padding)
                y2 = min(img.shape[0], y + h + padding)
                img = img[y1:y2, x1:x2]
            
            # Resize
            img = cv2.resize(img, target_size, interpolation=cv2.INTER_LANCZOS4)
            features.append(img)
        except Exception as e:
            print(f"Error processing {image_path}: {e}")
            continue
    
    return np.array(features)

print("Extracting features with face detection...")
X = extract_features_with_face_focus(df['image'], target_size=IMG_SIZE)
print(f"Feature shape: {X.shape}")

# Normalize to [-1, 1] for INT8 quantization
X = X.astype(np.float32) / 255.0
X = X * 2.0 - 1.0

# Labels
y_gender = np.array(df['gender'][:len(X)])

# Balance classes if needed
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_gender), y=y_gender)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"Class weights: {class_weight_dict}")

# Split with stratification
X_train, X_test, y_gender_train, y_gender_test = train_test_split(
    X, y_gender, test_size=0.2, random_state=42, stratify=y_gender
)

print(f"Training set: {X_train.shape}, Test set: {X_test.shape}")

# ===============================
# Optimized Model for Arduino (Lightweight)
# ===============================
def create_lightweight_gender_model(input_shape=(96, 96, 3)):
    """
    Lightweight model optimized for Arduino Nano 33 BLE
    Uses depthwise separable convolutions for efficiency
    """
    inputs = Input(shape=input_shape)
    
    # Block 1
    x = Conv2D(32, (3, 3), strides=(2, 2), padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    
    # Depthwise Separable Block 1
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(64, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    
    # Depthwise Separable Block 2
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(128, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    
    # Depthwise Separable Block 3
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(128, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    
    # Global Average Pooling
    x = GlobalAveragePooling2D()(x)
    
    # Dense layers
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    # Output
    outputs = Dense(1, activation='sigmoid', name='gender_output')(x)
    
    return Model(inputs, outputs)

# Create model
model = create_lightweight_gender_model(input_shape=(96, 96, 3))

# Compile with optimized settings
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)

model.summary()
total_params = model.count_params()
print(f"\nTotal Parameters: {total_params:,} ({total_params/1000:.1f}K)")

# ===============================
# Callbacks
# ===============================
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=12,
        restore_best_weights=True,
        mode='min',
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        mode='max',
        verbose=1
    ),
    ModelCheckpoint(
        'best_gender_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )
]

# ===============================
# Training with Data Augmentation
# ===============================
def augment_data(images, labels):
    """Apply data augmentation"""
    augmented_images = data_augmentation(images, training=True)
    return augmented_images, labels

print("\nStarting training...")
history = model.fit(
    X_train, y_gender_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_gender_test),
    callbacks=callbacks,
    class_weight=class_weight_dict,
    verbose=1
)

print(f"\nTraining completed! Total epochs run: {len(history.history['loss'])}")

# ===============================
# Evaluation
# ===============================
print("\nEvaluating model...")
loss, acc, precision, recall = model.evaluate(X_test, y_gender_test, verbose=0)
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"\nTest Metrics:")
print(f"Accuracy: {acc:.4f} ({acc*100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")

# ===============================
# Confusion Matrix
# ===============================
from sklearn.metrics import confusion_matrix, classification_report

y_pred = (model.predict(X_test, verbose=0) > 0.5).astype(int).flatten()
cm = confusion_matrix(y_gender_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Male', 'Female'], 
            yticklabels=['Male', 'Female'])
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

print("\nClassification Report:")
print(classification_report(y_gender_test, y_pred, target_names=['Male', 'Female']))

# ===============================
# Plot Training History
# ===============================
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Accuracy
axes[0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')
axes[0].set_title('Training & Validation Accuracy')
axes[0].legend()
axes[0].grid(True)

# Loss
axes[1].plot(history.history['loss'], label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Val Loss')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Loss')
axes[1].set_title('Training & Validation Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# ===============================
# Convert to TensorFlow Lite (INT8 Quantization)
# ===============================
print("\nConverting to TensorFlow Lite with INT8 quantization...")

def representative_dataset():
    """Representative dataset for quantization"""
    for i in range(min(100, len(X_train))):
        yield [X_train[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8
converter.representative_dataset = representative_dataset

tflite_model = converter.convert()

# Save the model
tflite_model_path = 'gender_model_96x96_int8.tflite'
with open(tflite_model_path, 'wb') as f:
    f.write(tflite_model)

print(f"TFLite model saved: {tflite_model_path}")
print(f"Model size: {len(tflite_model) / 1024:.2f} KB")

# ===============================
# Test TFLite Model
# ===============================
print("\nTesting TFLite model...")
interpreter = tf.lite.Interpreter(model_path=tflite_model_path)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print(f"Input shape: {input_details[0]['shape']}")
print(f"Input type: {input_details[0]['dtype']}")
print(f"Output shape: {output_details[0]['shape']}")
print(f"Output type: {output_details[0]['dtype']}")

# Test on sample images
correct_predictions = 0
total_samples = min(100, len(X_test))

for i in range(total_samples):
    # Quantize input
    input_scale, input_zero_point = input_details[0]['quantization']
    test_image = X_test[i:i+1].astype(np.float32)
    test_image_quantized = (test_image / input_scale + input_zero_point).astype(np.int8)
    
    # Run inference
    interpreter.set_tensor(input_details[0]['index'], test_image_quantized)
    interpreter.invoke()
    
    # Dequantize output
    output_data = interpreter.get_tensor(output_details[0]['index'])
    output_scale, output_zero_point = output_details[0]['quantization']
    output_dequantized = (output_data.astype(np.float32) - output_zero_point) * output_scale
    
    prediction = 1 if output_dequantized[0][0] >= 0.5 else 0
    if prediction == y_gender_test[i]:
        correct_predictions += 1

tflite_accuracy = correct_predictions / total_samples
print(f"TFLite Model Accuracy on {total_samples} samples: {tflite_accuracy:.4f} ({tflite_accuracy*100:.2f}%)")

# ===============================
# Convert to C Array for Arduino
# ===============================
print("\nGenerating C header file for Arduino...")

def convert_to_c_array(tflite_model, header_file_name):
    """Convert TFLite model to C array"""
    c_array = ','.join([str(byte) for byte in tflite_model])
    
    header_content = f"""// Auto-generated TensorFlow Lite model
// Model: Gender Classification (96x96, INT8 quantized)
// Generated on: {pd.Timestamp.now()}

#ifndef GENDER_MODEL_ARDUINO_H
#define GENDER_MODEL_ARDUINO_H

const unsigned char gender_model_arduino[] = {{
{c_array}
}};

const unsigned int gender_model_arduino_len = {len(tflite_model)};

#endif // GENDER_MODEL_ARDUINO_H
"""
    
    with open(header_file_name, 'w') as f:
        f.write(header_content)
    
    print(f"C header file saved: {header_file_name}")
    print(f"Array size: {len(tflite_model)} bytes")

convert_to_c_array(tflite_model, 'gender_model_arduino_96x96.h')

# ===============================
# Sample Predictions with Confidence
# ===============================
def test_predictions_with_confidence(model, X_test, y_gender_test, num_samples=10):
    indices = np.random.choice(len(X_test), num_samples, replace=False)
    fig, axes = plt.subplots(2, 5, figsize=(20, 8))
    axes = axes.flatten()
    
    for i, idx in enumerate(indices):
        pred = model.predict(X_test[idx:idx+1], verbose=0)
        confidence = pred[0][0]
        pred_gender = 1 if confidence >= 0.5 else 0
        
        # Denormalize image for display
        img_display = (X_test[idx] + 1.0) / 2.0
        
        axes[i].imshow(img_display)
        true_label = gender_dict[y_gender_test[idx]]
        pred_label = gender_dict[pred_gender]
        color = 'green' if pred_gender == y_gender_test[idx] else 'red'
        
        axes[i].set_title(f"True: {true_label}\nPred: {pred_label}\nConf: {confidence:.3f}", 
                         color=color, fontsize=10)
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

print("\nSample Predictions:")
test_predictions_with_confidence(model, X_test, y_gender_test)

# ===============================
# Final Summary
# ===============================
final_train_acc = max(history.history['accuracy'])
final_val_acc = max(history.history['val_accuracy'])

print(f"\n{'='*60}")
print(f"MODEL PERFORMANCE SUMMARY")
print(f"{'='*60}")
print(f"Total Parameters: {total_params:,} ({total_params/1000:.1f}K)")
print(f"Model Size (TFLite): {len(tflite_model) / 1024:.2f} KB")
print(f"Input Shape: 96x96x3 (INT8)")
print(f"Best Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
print(f"Best Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
print(f"Test Accuracy (Keras): {acc:.4f} ({acc*100:.2f}%)")
print(f"Test Accuracy (TFLite): {tflite_accuracy:.4f} ({tflite_accuracy*100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1_score:.4f}")
print(f"{'='*60}")

print("\n✅ Training complete!")
print("📁 Files generated:")
print("   - gender_model_96x96_int8.tflite")
print("   - gender_model_arduino_96x96.h")
print("   - best_gender_model.h5")
print("\n🚀 Next steps:")
print("   1. Copy 'gender_model_arduino_96x96.h' to your Arduino project")
print("   2. Upload the improved Arduino code")
print("   3. Test with face images in front of camera")
```
