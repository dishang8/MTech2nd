```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from tqdm.notebook import tqdm
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Dense, Conv2D, Dropout, MaxPooling2D, 
                                      Input, BatchNormalization, DepthwiseConv2D, 
                                      GlobalAveragePooling2D, Activation, Flatten)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from PIL import Image
import cv2

# ===============================
# Configuration - OPTIMIZED 64x64
# ===============================
BASE_DIR = r"C:\Users\Dishang\OneDrive\Desktop\UTKFace"
IMG_SIZE = (64, 64)  # Sweet spot between 48 and 96
BATCH_SIZE = 32
EPOCHS = 60

print(f"🎯 Training OPTIMIZED 64×64 Gender Classification Model")
print(f"   Target: Good accuracy + Arduino compatible")
print(f"   Expected RAM: ~140-160KB")

# ===============================
# Strict Face Validation
# ===============================
def validate_face_image(image_path):
    """Strict validation - must have clear, centered face"""
    try:
        img = cv2.imread(image_path)
        if img is None:
            return False, None
        
        h, w = img.shape[:2]
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # Use multiple cascade classifiers for better detection
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
        
        faces = face_cascade.detectMultiScale(gray, 1.1, 5, minSize=(30, 30))
        
        if len(faces) == 0:
            return False, None
        
        # Get largest face
        x, y, fw, fh = max(faces, key=lambda f: f[2] * f[3])
        
        # Check if face is reasonably centered
        face_center_x = x + fw/2
        face_center_y = y + fh/2
        img_center_x = w/2
        img_center_y = h/2
        
        # Face should be within 40% of image center
        if (abs(face_center_x - img_center_x) > w*0.4 or 
            abs(face_center_y - img_center_y) > h*0.4):
            return False, None
        
        # Face should be at least 30% of image
        if (fw * fh) < (w * h * 0.09):  # 30% width * 30% height = 9% area
            return False, None
        
        # Check for eyes (better face validation)
        face_roi = gray[y:y+fh, x:x+fw]
        eyes = eye_cascade.detectMultiScale(face_roi, 1.1, 3)
        
        # Should have at least 1 eye detected
        if len(eyes) < 1:
            return False, None
        
        return True, (x, y, fw, fh)
    
    except:
        return False, None

# ===============================
# Load Dataset with STRICT validation
# ===============================
print("Loading dataset with STRICT face validation...")
image_paths = []
age_labels = []
gender_labels = []
face_boxes = []

skipped_no_face = 0
skipped_off_center = 0
skipped_too_small = 0
skipped_no_eyes = 0

for filename in tqdm(os.listdir(BASE_DIR)):
    try:
        image_path = os.path.join(BASE_DIR, filename)
        temp = filename.split('_')
        age = int(temp[0])
        gender = int(temp[1])
        
        is_valid, face_box = validate_face_image(image_path)
        
        if is_valid:
            image_paths.append(image_path)
            age_labels.append(age)
            gender_labels.append(gender)
            face_boxes.append(face_box)
        else:
            skipped_no_face += 1
            
    except:
        continue

df = pd.DataFrame()
df['image'] = image_paths
df['age'] = age_labels
df['gender'] = gender_labels
df['face_box'] = face_boxes

print(f"\n✓ Valid images: {len(df)}")
print(f"✗ Skipped (no clear face): {skipped_no_face}")
print(f"\nGender distribution:")
print(df['gender'].value_counts())
print(f"Balance ratio: {df['gender'].value_counts().min() / df['gender'].value_counts().max():.2f}")

gender_dict = {0: 'Male', 1: 'Female'}

# ===============================
# Enhanced Feature Extraction
# ===============================
def extract_features_with_augmentation(images, face_boxes, target_size=IMG_SIZE):
    """Extract with tight face cropping and augmentation awareness"""
    features = []
    
    for idx, image_path in enumerate(tqdm(images)):
        try:
            img = cv2.imread(image_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            # Use detected face box for tight crop
            if face_boxes[idx] is not None:
                x, y, w, h = face_boxes[idx]
                # Add small padding (10%)
                padding = int(0.1 * max(w, h))
                x1 = max(0, x - padding)
                y1 = max(0, y - padding)
                x2 = min(img.shape[1], x + w + padding)
                y2 = min(img.shape[0], y + h + padding)
                img = img[y1:y2, x1:x2]
            
            # Resize with high quality
            img = cv2.resize(img, target_size, interpolation=cv2.INTER_LANCZOS4)
            
            # Histogram equalization for better contrast
            img_yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
            img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
            img = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
            
            features.append(img)
        except Exception as e:
            print(f"Error processing {image_path}: {e}")
            continue
    
    return np.array(features)

print("\nExtracting features with tight face cropping...")
X = extract_features_with_augmentation(df['image'].values, df['face_box'].values, target_size=IMG_SIZE)
print(f"✓ Feature shape: {X.shape}")

# Normalize to [-1, 1]
X = X.astype(np.float32) / 255.0
X = X * 2.0 - 1.0

y_gender = np.array(df['gender'][:len(X)])

# Compute class weights
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight('balanced', classes=np.unique(y_gender), y=y_gender)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
print(f"✓ Class weights: {class_weight_dict}")

# Stratified split
X_train, X_test, y_gender_train, y_gender_test = train_test_split(
    X, y_gender, test_size=0.2, random_state=42, stratify=y_gender
)

print(f"✓ Training: {X_train.shape}, Test: {X_test.shape}")

# ===============================
# Data Augmentation
# ===============================
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.08),
    tf.keras.layers.RandomZoom(0.08),
    tf.keras.layers.RandomContrast(0.15),
], name='augmentation')

# ===============================
# Optimized 64×64 Model
# ===============================
def create_optimized_64x64_model(input_shape=(64, 64, 3)):
    """
    Balanced model for 64×64 input
    Target: ~35-45K parameters, ~90-110KB size, ~140-160KB RAM
    """
    inputs = Input(shape=input_shape, name='input')
    
    # Block 1: Initial conv
    x = Conv2D(24, (3, 3), strides=(2, 2), padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    # 32×32×24
    
    # Block 2: Depthwise separable
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(48, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    # 16×16×48
    
    # Block 3: Depthwise separable
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(96, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2, 2))(x)
    # 8×8×96
    
    # Block 4: One more depthwise
    x = DepthwiseConv2D((3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Conv2D(128, (1, 1), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    # 8×8×128
    
    # Global pooling
    x = GlobalAveragePooling2D()(x)
    # 128
    
    # Dense layers
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(32, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    # Output
    outputs = Dense(1, activation='sigmoid', name='gender_output')(x)
    
    return Model(inputs, outputs, name='Gender64x64Model')

# Create and compile
model = create_optimized_64x64_model()

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', 
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall')]
)

model.summary()
total_params = model.count_params()
print(f"\n✓ Total Parameters: {total_params:,} (~{total_params/1000:.1f}K)")

# ===============================
# Callbacks
# ===============================
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=12,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_gender_64x64_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# ===============================
# Training
# ===============================
print("\n🚀 Starting training...")
history = model.fit(
    X_train, y_gender_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_gender_test),
    callbacks=callbacks,
    class_weight=class_weight_dict,
    verbose=1
)

print(f"\n✓ Training completed!")

# ===============================
# Evaluation
# ===============================
results = model.evaluate(X_test, y_gender_test, verbose=0)
loss = results[0]
acc = results[1]
precision = results[2]
recall = results[3]
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print(f"\n{'='*50}")
print(f"KERAS MODEL PERFORMANCE")
print(f"{'='*50}")
print(f"Loss:      {loss:.4f}")
print(f"Accuracy:  {acc:.4f} ({acc*100:.2f}%)")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1_score:.4f}")
print(f"{'='*50}")

# ===============================
# Confusion Matrix
# ===============================
from sklearn.metrics import confusion_matrix, classification_report

y_pred = (model.predict(X_test, verbose=0) > 0.5).astype(int).flatten()
cm = confusion_matrix(y_gender_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Male', 'Female'],
            yticklabels=['Male', 'Female'])
plt.title('Confusion Matrix - 64×64 Model')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

print("\nDetailed Classification Report:")
print(classification_report(y_gender_test, y_pred, target_names=['Male', 'Female']))

# ===============================
# TFLite Conversion (INT8)
# ===============================
print("\n🔄 Converting to TFLite INT8...")

def representative_dataset():
    for i in range(min(200, len(X_train))):
        yield [X_train[i:i+1].astype(np.float32)]

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8
converter.representative_dataset = representative_dataset

tflite_model = converter.convert()

tflite_path = 'gender_model_64x64_int8.tflite'
with open(tflite_path, 'wb') as f:
    f.write(tflite_model)

model_size_kb = len(tflite_model) / 1024
print(f"✓ TFLite model saved: {tflite_path}")
print(f"✓ Model size: {model_size_kb:.2f} KB")

# ===============================
# Test TFLite
# ===============================
print("\n📊 Testing TFLite model...")
interpreter = tf.lite.Interpreter(model_path=tflite_path)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print(f"\nTFLite Model Info:")
print(f"  Input shape: {input_details[0]['shape']}")
print(f"  Input type: {input_details[0]['dtype']}")
print(f"  Output shape: {output_details[0]['shape']}")
print(f"  Output type: {output_details[0]['dtype']}")

# Test accuracy
correct = 0
total = min(200, len(X_test))

for i in range(total):
    input_scale, input_zero_point = input_details[0]['quantization']
    test_image = X_test[i:i+1].astype(np.float32)
    test_image_quantized = (test_image / input_scale + input_zero_point).astype(np.int8)
    
    interpreter.set_tensor(input_details[0]['index'], test_image_quantized)
    interpreter.invoke()
    
    output_data = interpreter.get_tensor(output_details[0]['index'])
    output_scale, output_zero_point = output_details[0]['quantization']
    output_dequantized = (output_data.astype(np.float32) - output_zero_point) * output_scale
    
    prediction = 1 if output_dequantized[0][0] >= 0.5 else 0
    if prediction == y_gender_test[i]:
        correct += 1

tflite_acc = correct / total
print(f"✓ TFLite Accuracy ({total} samples): {tflite_acc:.4f} ({tflite_acc*100:.2f}%)")

# ===============================
# Arduino RAM Estimation
# ===============================
input_size = np.prod(input_details[0]['shape'])  # 64*64*3 = 12,288
estimated_arena = model_size_kb * 1.3

print(f"\n{'='*50}")
print(f"ARDUINO COMPATIBILITY")
print(f"{'='*50}")
print(f"Input tensor:    {input_size:,} bytes ({input_size/1024:.1f} KB)")
print(f"Model size:      {model_size_kb:.2f} KB")
print(f"Est. arena:      {estimated_arena:.0f} KB")
print(f"Total est.:      {(estimated_arena + input_size/1024):.0f} KB")
print(f"Arduino RAM:     256 KB")
print(f"Status:          {'✅ FITS' if (estimated_arena + input_size/1024) < 200 else '⚠️ TIGHT'}")
print(f"{'='*50}")

# ===============================
# Generate C Header
# ===============================
print("\n📝 Generating C header...")

header_content = f"""// Auto-generated TensorFlow Lite Model
// Model: Gender Classification 64×64 INT8
// Generated: {pd.Timestamp.now()}
// Size: {len(tflite_model)} bytes ({model_size_kb:.2f} KB)
// Parameters: {total_params:,}
// Accuracy: {acc*100:.1f}% (Keras), {tflite_acc*100:.1f}% (TFLite)

#ifndef GENDER_MODEL_ARDUINO_64X64_H
#define GENDER_MODEL_ARDUINO_64X64_H

const unsigned char gender_model_arduino[] = {{
{','.join(str(b) for b in tflite_model)}
}};

const unsigned int gender_model_arduino_len = {len(tflite_model)};

#endif // GENDER_MODEL_ARDUINO_64X64_H
"""

header_file = 'gender_model_arduino_64x64.h'
with open(header_file, 'w') as f:
    f.write(header_content)

print(f"✓ Header file: {header_file}")

# ===============================
# Plot History
# ===============================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history.history['accuracy'], label='Train', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Validation', linewidth=2)
axes[0].set_xlabel('Epoch', fontsize=12)
axes[0].set_ylabel('Accuracy', fontsize=12)
axes[0].set_title('Training Accuracy', fontsize=14, fontweight='bold')
axes[0].legend(fontsize=11)
axes[0].grid(True, alpha=0.3)

axes[1].plot(history.history['loss'], label='Train', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Validation', linewidth=2)
axes[1].set_xlabel('Epoch', fontsize=12)
axes[1].set_ylabel('Loss', fontsize=12)
axes[1].set_title('Training Loss', fontsize=14, fontweight='bold')
axes[1].legend(fontsize=11)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ===============================
# Final Summary
# ===============================
print(f"\n{'='*60}")
print(f"✅ TRAINING COMPLETE - 64×64 OPTIMIZED MODEL")
print(f"{'='*60}")
print(f"Model Size:          {model_size_kb:.2f} KB")
print(f"Parameters:          {total_params:,}")
print(f"Input Shape:         64×64×3 (INT8)")
print(f"Keras Accuracy:      {acc:.4f} ({acc*100:.2f}%)")
print(f"TFLite Accuracy:     {tflite_acc:.4f} ({tflite_acc*100:.2f}%)")
print(f"Precision:           {precision:.4f}")
print(f"Recall:              {recall:.4f}")
print(f"F1-Score:            {f1_score:.4f}")
print(f"Est. Arduino RAM:    ~{(estimated_arena + input_size/1024):.0f} KB")
print(f"{'='*60}")
print(f"\n📁 Files generated:")
print(f"   ✓ {tflite_path}")
print(f"   ✓ {header_file}")
print(f"   ✓ best_gender_64x64_model.h5")
print(f"\n🚀 Next: Update Arduino code and test!")
```
