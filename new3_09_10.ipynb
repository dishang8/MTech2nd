```python
          if grayscale:
                img = np.expand_dims(img, axis=-1)  # Add channel dimension
            features.append(img)
        except:
            continue
    features = np.array(features)
    return features

X = extract_features_arduino(df['image'], target_size=(IMG_SIZE, IMG_SIZE), grayscale=GRAYSCALE)
print(f"\nFeature shape: {X.shape}")

# Normalize to [-1, 1] for better quantization
X = (X / 127.5) - 1.0

# Labels
y_gender = np.array(df['gender'][:len(X)])

# Split with stratification
X_train, X_test, y_gender_train, y_gender_test = train_test_split(
    X, y_gender, test_size=0.2, random_state=42, stratify=y_gender
)

print(f"Training set: {X_train.shape}, Test set: {X_test.shape}")

# ===============================
# Ultra-Lightweight Model for Arduino
# ===============================
def create_arduino_model(input_shape):
    """
    Ultra-lightweight model optimized for Arduino Nano 33 BLE
    Target: <20KB model size, <10K parameters
    Uses MobileNet-style depthwise separable convolutions
    """
    inputs = Input(input_shape)
    
    # Block 1: Initial feature extraction
    x = Conv2D(8, (3, 3), strides=2, padding='same', activation='relu')(inputs)
    x = BatchNormalization()(x)
    
    # Block 2: Depthwise separable conv (memory efficient)
    x = DepthwiseConv2D((3, 3), padding='same', activation='relu')(x)
    x = Conv2D(16, (1, 1), padding='same', activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    
    # Block 3: Another depthwise block
    x = DepthwiseConv2D((3, 3), padding='same', activation='relu')(x)
    x = Conv2D(32, (1, 1), padding='same', activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2))(x)
    
    # Global pooling instead of flatten (reduces parameters)
    x = GlobalAveragePooling2D()(x)
    
    # Minimal dense layer
    x = Dense(16, activation='relu')(x)
    x = Dropout(0.3)(x)
    
    # Output
    outputs = Dense(1, activation='sigmoid', name='gender')(x)
    
    return Model(inputs, outputs)

# Create model
input_shape = (IMG_SIZE, IMG_SIZE, 1 if GRAYSCALE else 3)
model = create_arduino_model(input_shape)

# Compile
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
total_params = model.count_params()
print(f"\n{'='*60}")
print(f"Total Parameters: {total_params:,}")
print(f"Estimated float32 model size: ~{total_params*4/1024:.1f} KB")
print(f"Estimated quantized model size: ~{total_params/1024:.1f} KB")
print(f"{'='*60}")

# ===============================
# Callbacks
# ===============================
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        mode='min',
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        mode='max',
        verbose=1
    )
]

# ===============================
# Training
# ===============================
print("\nStarting training...")
history = model.fit(
    X_train, y_gender_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(X_test, y_gender_test),
    callbacks=callbacks,
    verbose=1
)

print(f"\nTraining completed! Total epochs run: {len(history.history['loss'])}")

# ===============================
# Evaluation
# ===============================
loss, acc = model.evaluate(X_test, y_gender_test, verbose=0)
print(f"\nTest Accuracy: {acc:.4f} ({acc*100:.2f}%)")

# ===============================
# Convert to TensorFlow Lite with Quantization
# ===============================
print("\n" + "="*60)
print("CONVERTING TO TENSORFLOW LITE")
print("="*60)

# Save Keras model first
model.save('gender_model_arduino.h5')
print("✓ Keras model saved: gender_model_arduino.h5")

# Convert to TFLite with full integer quantization
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Quantization for Arduino (INT8)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

# Representative dataset for quantization
def representative_dataset():
    for i in range(100):
        yield [X_train[i:i+1].astype(np.float32)]

converter.representative_dataset = representative_dataset

# Convert
tflite_model = converter.convert()

# Save TFLite model
tflite_path = 'gender_model_arduino.tflite'
with open(tflite_path, 'wb') as f:
    f.write(tflite_model)

model_size_kb = len(tflite_model) / 1024
print(f"✓ TFLite model saved: {tflite_path}")
print(f"✓ Model size: {model_size_kb:.2f} KB")

if model_size_kb < 20:
    print(f"✓ SUCCESS: Model fits Arduino Nano 33 BLE! ({model_size_kb:.2f} KB < 20 KB)")
else:
    print(f"⚠ WARNING: Model might be too large ({model_size_kb:.2f} KB)")

# ===============================
# Test TFLite Model
# ===============================
print("\nTesting TFLite model accuracy...")
interpreter = tf.lite.Interpreter(model_path=tflite_path)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print(f"Input shape: {input_details[0]['shape']}")
print(f"Input type: {input_details[0]['dtype']}")
print(f"Output shape: {output_details[0]['shape']}")
print(f"Output type: {output_details[0]['dtype']}")

# Test on sample
correct = 0
total = min(100, len(X_test))

for i in range(total):
    # Prepare input (quantize to int8)
    input_data = X_test[i:i+1].astype(np.float32)
    input_scale, input_zero_point = input_details[0]['quantization']
    input_data = (input_data / input_scale + input_zero_point).astype(np.int8)
    
    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()
    
    # Get output (dequantize from int8)
    output_data = interpreter.get_tensor(output_details[0]['index'])
    output_scale, output_zero_point = output_details[0]['quantization']
    output_data = (output_data.astype(np.float32) - output_zero_point) * output_scale
    
    pred = 1 if output_data[0][0] > 0.5 else 0
    if pred == y_gender_test[i]:
        correct += 1

tflite_accuracy = correct / total
print(f"TFLite model accuracy (sample): {tflite_accuracy:.4f} ({tflite_accuracy*100:.2f}%)")

# ===============================
# Generate Arduino Header File
# ===============================
def generate_arduino_header(tflite_model, filename='model_data.h'):
    with open(filename, 'w') as f:
        f.write("// Auto-generated model data for Arduino Nano 33 BLE\n")
        f.write("// Gender Classification Model\n\n")
        f.write("#ifndef MODEL_DATA_H\n")
        f.write("#define MODEL_DATA_H\n\n")
        f.write(f"const unsigned int model_len = {len(tflite_model)};\n")
        f.write("const unsigned char model_data[] = {\n")
        
        hex_array = [f"0x{b:02x}" for b in tflite_model]
        for i in range(0, len(hex_array), 12):
            f.write("  " + ", ".join(hex_array[i:i+12]) + ",\n")
        
        f.write("};\n\n")
        f.write("#endif // MODEL_DATA_H\n")
    
    print(f"✓ Arduino header file generated: {filename}")

generate_arduino_header(tflite_model)

# ===============================
# Plot Training History
# ===============================
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title("Training & Validation Accuracy")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Training & Validation Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ===============================
# Sample Predictions
# ===============================
def test_predictions(model, X_test, y_gender_test, num_samples=5):
    indices = np.random.choice(len(X_test), num_samples, replace=False)
    plt.figure(figsize=(15, 3))

    for i, idx in enumerate(indices):
        pred = model.predict(X_test[idx:idx+1], verbose=0)
        pred_gender = round(pred[0][0])
        conf = pred[0][0]

        plt.subplot(1, num_samples, i+1)
        if GRAYSCALE:
            plt.imshow(X_test[idx].squeeze(), cmap='gray')
        else:
            plt.imshow((X_test[idx] + 1.0) / 2.0)  # Denormalize
        plt.title(f"True: {gender_dict[y_gender_test[idx]]}\nPred: {gender_dict[pred_gender]}\nConf: {conf:.2f}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

print("\nSample Predictions:")
test_predictions(model, X_test, y_gender_test)

# ===============================
# Final Summary
# ===============================
final_train_acc = max(history.history['accuracy'])
final_val_acc = max(history.history['val_accuracy'])

print(f"\n{'='*60}")
print(f"FINAL MODEL SUMMARY - ARDUINO NANO 33 BLE")
print(f"{'='*60}")
print(f"Input Size: {IMG_SIZE}x{IMG_SIZE} {'Grayscale' if GRAYSCALE else 'RGB'}")
print(f"Total Parameters: {total_params:,}")
print(f"Float32 Model Size: ~{total_params*4/1024:.1f} KB")
print(f"Quantized Model Size: {model_size_kb:.2f} KB")
print(f"Best Training Accuracy: {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
print(f"Best Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
print(f"Test Accuracy: {acc:.4f} ({acc*100:.2f}%)")
print(f"TFLite Accuracy: {tflite_accuracy:.4f} ({tflite_accuracy*100:.2f}%)")
print(f"{'='*60}")
print(f"\nFiles generated:")
print(f"  1. gender_model_arduino.h5 - Keras model")
print(f"  2. gender_model_arduino.tflite - TFLite model for Arduino")
print(f"  3. model_data.h - Arduino header file")
print(f"{'='*60}")
```

```python
import cv2
import numpy as np
import tensorflow as tf

# Load TFLite gender model
interpreter = tf.lite.Interpreter(model_path="gender_model_arduino.tflite")
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

input_shape = input_details[0]['shape']
input_dtype = input_details[0]['dtype']
IMG_SIZE = 48

# Load OpenCV Haar Cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

def preprocess_face(face_img):
    gray = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (IMG_SIZE, IMG_SIZE))
    img = (resized / 127.5) - 1.0
    img = np.expand_dims(img, axis=(0, -1)).astype(np.float32)

    # Quantize if model is int8
    if input_dtype == np.int8:
        scale, zero_point = input_details[0]['quantization']
        img = (img / scale + zero_point).astype(np.int8)
    return img

# Open webcam
path = r"C:\Users\Dishang\OneDrive\Desktop\test_video.webm"
cap = cv2.VideoCapture(path)
print("Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detect faces
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5, minSize=(60, 60))

    if len(faces) == 0:
        cv2.putText(frame, "No face detected", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    else:
        for (x, y, w, h) in faces:
            face_roi = frame[y:y+h, x:x+w]
            input_data = preprocess_face(face_roi)

            interpreter.set_tensor(input_details[0]['index'], input_data)
            interpreter.invoke()
            output_data = interpreter.get_tensor(output_details[0]['index'])

            # Dequantize output
            if output_details[0]['dtype'] == np.int8:
                scale, zero_point = output_details[0]['quantization']
                output_data = (output_data.astype(np.float32) - zero_point) * scale

            confidence = float(output_data[0][0])
            gender = "Female" if confidence >= 0.5 else "Male"
            color = (255, 0, 255) if gender == "Female" else (0, 255, 255)

            # Draw bounding box and label
            cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
            cv2.putText(frame, f"{gender} ({confidence:.2f})", (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
            

    cv2.imshow("Gender Detection (with Face)", frame)

    if cv2.waitKey(1000) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
